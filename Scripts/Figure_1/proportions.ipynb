```python
from Bio import AlignIO
from Bio import SeqIO
from Bio.Seq import Seq
import csv
import baltic as bt
import collections
import math
import collections
import os
import csv
from Bio import SeqIO
import os
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import collections
import numpy as np
import pandas as pd
import datetime
from datetime import date
import skbio
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.path as mpath
import matplotlib.lines as mlines
import collections
import matplotlib.patches as patches
import matplotlib.colors as mcolors
from matplotlib.patches import Polygon
from matplotlib import cm

type_counter = collections.Counter()
apobec_counter = collections.Counter()

df=pd.read_csv("ACEGID.aln.fasta.pruned.treefile.branch_snps.reconstruction_cladeedits.csv")

hdf=df[df["Clade"]=="Clade2B"]
hdf["APOBEC"]="Non-Apobec"
ii=hdf[(hdf["snp"].isin(["C->T", "G->A"])) & (hdf["dimer"].isin(["TC","GA"]))].index
hdf["APOBEC"].loc[ii]="APOBEC"

df1=pd.DataFrame(hdf.groupby("snp")["APOBEC"].value_counts())
df1["Clade"]="Clade2B"

hdf=df[df["Clade"]=="Zoonotic"]
hdf["APOBEC"]="Non-Apobec"
ii=hdf[(hdf["snp"].isin(["C->T", "G->A"])) & (hdf["dimer"].isin(["TC","GA"]))].index
hdf["APOBEC"].loc[ii]="APOBEC"

df2=pd.DataFrame(hdf.groupby("snp")["APOBEC"].value_counts())
df2["Clade"]="Zoonotic"

hdf=df[df["Clade"]=="LinC"]
hdf["APOBEC"]="Non-Apobec"
ii=hdf[(hdf["snp"].isin(["C->T", "G->A"])) & (hdf["dimer"].isin(["TC","GA"]))].index
hdf["APOBEC"].loc[ii]="APOBEC"

df3=pd.DataFrame(hdf.groupby("snp")["APOBEC"].value_counts())
df3["Clade"]="LinC"


MT = pd.concat([df1, df2, df3])
MT=MT.reset_index()

snps=list(MT["snp"].unique())

print(len(MT))
for snp in snps:
    for clade in ["Clade2B", "Zoonotic", "LinC"]:
        
        for apo in ["Non-Apobec", "APOBEC"]:
            
            if len(MT[(MT["snp"]==snp) & (MT["Clade"]==clade) & (MT["APOBEC"]==apo)]) <1:
                
                ndf=pd.DataFrame([snp, apo, 0, clade])
                ndf=ndf.T.rename(columns={0:"snp", 1:"APOBEC", 2:"count", 3:"Clade"})
                # print(clade)
                # print(ndf)
                MT=pd.concat([MT, ndf], axis=0)


MT=MT.sort_values(["Clade", "snp", "APOBEC"])
o=MT[(MT["Clade"]=="Clade2B") & (MT["APOBEC"]=="APOBEC")].sort_values("count", ascending=False)["snp"].to_list()
MT['snp'] = pd.Categorical(MT['snp'], categories=o, ordered=True)
MT = MT.sort_values(by='snp')
MT.to_csv("SNP_counts.csv", index="None")
mpl.rcParams['font.family'] = 'sans-serif'


fig, ax = plt.subplots(2, 1, figsize=(5,8))
axnr=0
lbz=["hMPX1", "Zoonotic"]

handles,labels=[],[]

for c in ["Clade2B", "Zoonotic"]:
    ax[axnr].bar(MT[(MT["Clade"]==c) & (MT["APOBEC"]=="APOBEC")]["snp"], MT[(MT["Clade"]==c) & (MT["APOBEC"]=="APOBEC") ]["count"],alpha=1, color="#B31B1B", edgecolor='black')
    ax[axnr].bar(MT[(MT["Clade"]==c) & (MT["APOBEC"]=="Non-Apobec")]["snp"], MT[(MT["Clade"]==c) & (MT["APOBEC"]=="Non-Apobec") ]["count"],
                 bottom= MT[(MT["Clade"]==c) & (MT["APOBEC"]=="APOBEC") ]["count"], color="#BEBFC5", alpha=1, edgecolor='black')
    ax[axnr].set_title(label=lbz[axnr], fontsize=30)
    ax[axnr].tick_params(axis='x', which='major', labelsize=16, rotation=90)
    ax[axnr].set_ylabel('Count', fontsize=30)
    ax[axnr].tick_params(axis='y', which='major', labelsize=16)
    ax[axnr].spines['right'].set_visible('off')
    ax[axnr].spines['top'].set_visible('off')
    axnr=axnr+1

ax[0].scatter(5,400,s=400,color='#B31B1B',zorder=20, edgecolor='black')
ax[0].scatter(5,340,s=400,color='#BEBFC5',zorder=20, edgecolor='black')
ax[0].text(5.5, 380, "APOBEC", size=20)
ax[0].text(5.5, 320, "Non-APOBEC", size=20)


ax[0].spines['right'].set_visible(False)
ax[0].spines['top'].set_visible(False)


ax[1].spines['right'].set_visible(False)
ax[1].spines['top'].set_visible(False)

# plt.legend()
plt.tight_layout()



plt.savefig("Internal_branches.png",bbox_inches='tight', dpi=1000)
```

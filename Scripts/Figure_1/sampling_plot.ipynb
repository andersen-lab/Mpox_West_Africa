```python
import baltic as bt
import requests
from io import StringIO as sio
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from matplotlib.patches import Polygon, PathPatch
from matplotlib.path import Path
from matplotlib.collections import LineCollection
from matplotlib import cm
import matplotlib.patheffects as path_effects
from IPython.display import HTML
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime as dt
import os
import PyAstronomy
from PyAstronomy import pyasl
import datetime
import re
import matplotlib as mpl
from matplotlib import pyplot as plt
import matplotlib.patheffects as path_effects
from matplotlib.gridspec import GridSpec
from matplotlib.patches import Polygon
import pandas as pd

mpl.rcParams['font.family'] = 'sans-serif'

dd={'north_central': "#960018",
    'north_east': "#FE5A1D",
     'north_west': "#FFD700",
      'south': "#16166B",
    'south_east': "#0070BB",
     'south_west': "#138808",
           'Not_specified': "#C0C0C0"}

dl={'north_central': "North Central",
    'north_east': "North East",
     'north_west': "North West",
      'south': "South South",
    'south_east': "South East",
     'south_west': "South West",
           'Not_specified': "NA"}


rn={'NC':'north_central',
    'NE':'north_east',
     'NW':'north_west',
      'S':'south',
    'SE':'south_east',
     'SE':'south_west'}


df=pd.read_csv("epidata_region.csv")

pivot_df = df.pivot_table(index='ymd', columns='Region', values='n', fill_value=0)
pivot_df.reset_index(inplace=True)
melted_df = pd.melt(pivot_df, id_vars='ymd', var_name='Region', value_name='n')
date_list = [re.sub(" (.*)", "", str(x)) for x in pd.date_range(start='2017-06-15', end='2023-05-15', freq='MS') + pd.Timedelta(days=14)]
melted_df["ymd"]=melted_df["ymd"].astype(str)


for r in melted_df["Region"].unique():
    missing=[x for x in date_list if x not in melted_df[melted_df["Region"]==r]["ymd"].to_list()]
    for d in missing:
        new_row=[d, r, 0]
        melted_df.loc[len(melted_df)] = new_row


melted_df["ymd"]=melted_df["ymd"].str.replace("-15", "")
melted_df["ymd"]=pd.to_datetime(melted_df["ymd"])
melted_df["ymd"]=melted_df["ymd"].dt.strftime('%Y-%m')
melted_df=melted_df.sort_values(["Region", "ymd"])

MT = pd.DataFrame(columns=['ymd', 'Region', 'n'])

for r in melted_df["Region"].unique():
    for d in melted_df[melted_df["Region"]==r]["ymd"]:
        d_f=re.sub(" (.*)", "", str(d))
        y=str(d_f.split("-")[0])
        m=str(d_f.split("-")[1])
        val=melted_df[(melted_df["Region"]==r) & (melted_df["ymd"]==d)]["n"].values[0]
        for dayd in range(1, 28, 1): 
            if dayd < 10:
                new_row=[y+"-"+m+"-0"+str(dayd), r, val]
                MT.loc[len(MT)] = new_row
            else:
                new_row=[y+"-"+m+"-"+str(dayd), r, val]
                MT.loc[len(MT)] = new_row        


```

```python
gdf=pd.read_csv("GEO_states.csv")

gdf=gdf[(gdf["Taxa"].str.contains("VSP")) | (gdf["Taxa"].str.contains("TRM"))]
gdf["Region"]=gdf["Region"].str.replace("nigeria", "S")
gdf["Date"]=[x.split("_")[-1] for x in gdf["Taxa"]]
gdf["DD"]=["-".join(x.split("-")[0:2]) for x in gdf["Date"].to_list()]
gdf["DD"].loc[gdf["DD"].str.count("-")< 1]=gdf.loc[gdf["DD"].str.count("-")< 1]["DD"] + "-06"
gdf=gdf.groupby(['Region', 'DD']).size().reset_index(name='Count')
gdf=gdf.rename(columns={"Region":"Region", "DD":"ymd", "Count":"n"})

gdf["ymd"]=[x+"-15" for x in gdf["ymd"].to_list()]
```
